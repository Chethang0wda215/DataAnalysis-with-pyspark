```python
!unzip archive.zip -d taxi_data/     # Extract to folder
df = spark.read.csv("taxi_data/*.csv", header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import col   # change to correct data type

df_typed = df \
    .withColumn("trip_distance", col("trip_distance").cast("double")) \
    .withColumn("fare_amount", col("fare_amount").cast("double")) \
    .withColumn("extra", col("extra").cast("double")) \
    .withColumn("mta_tax", col("mta_tax").cast("double")) \
    .withColumn("tip_amount", col("tip_amount").cast("double")) \
    .withColumn("tolls_amount", col("tolls_amount").cast("double")) \
    .withColumn("imp_surcharge", col("imp_surcharge").cast("double")) \
    .withColumn("total_amount", col("total_amount").cast("double")) \
    .withColumn("year", col("year").cast("int")) \
    .withColumn("month", col("month").cast("int")) \
    .withColumn("day", col("day").cast("int"))
```

```python
df_typed.printSchema()
```

```python
df_clean = df_typed.filter(     #Remove invalid or corrupt records
    (col("trip_distance") > 0) &
    (col("fare_amount") > 0) &
    (col("total_amount") > 0)
)
```

```python
df_clean = df_clean.dropna(subset=[    # claean null
    "pickup_location_id",
    "dropoff_location_id",
    "hour_of_day"
])
```

```python
df_features = df_clean.withColumn(      #Feature Engineering (Big Data Friendly)1️⃣ Trip speed (distance per hour)
    "trip_speed",
    col("trip_distance") / (col("trip_duration") / 3600)
)
```

```python
df_features = df_features.withColumn(       #Tip percentage
    "tip_percentage",
    (col("tip_amount") / col("fare_amount")) * 100
)
```

```python
trips_by_hour = (
    df_features
    .groupBy("hour_of_day")
    .count()
    .orderBy("hour_of_day")
)

```

```python
df = spark.read.csv(
    "taxi_data/taxi_trip_data.csv",
    header=True,
    inferSchema=False
)
```

```python
from pyspark.sql.functions import expr, col
```

```python
from pyspark.sql.functions import col, expr
```

```python
df.columns
```

```python
from pyspark.sql.functions import to_timestamp

df_time = df.withColumn(
    "pickup_datetime",
    to_timestamp("pickup_datetime", "yyyy-MM-dd HH:mm:ss")
)
```

```python
df_time.printSchema()
```

```python
from pyspark.sql.functions import year, month, dayofmonth, dayofweek, hour

df_time = df_time \
    .withColumn("year", year("pickup_datetime")) \
    .withColumn("month", month("pickup_datetime")) \
    .withColumn("day", dayofmonth("pickup_datetime")) \
    .withColumn("day_of_week", dayofweek("pickup_datetime")) \
    .withColumn("hour_of_day", hour("pickup_datetime"))
```

```python
from pyspark.sql.functions import expr

df_typed = df_time \
    .withColumn("trip_distance", expr("try_cast(trip_distance as double)")) \
    .withColumn("fare_amount", expr("try_cast(fare_amount as double)")) \
    .withColumn("extra", expr("try_cast(extra as double)")) \
    .withColumn("mta_tax", expr("try_cast(mta_tax as double)")) \
    .withColumn("tip_amount", expr("try_cast(tip_amount as double)")) \
    .withColumn("tolls_amount", expr("try_cast(tolls_amount as double)")) \
    .withColumn("imp_surcharge", expr("try_cast(imp_surcharge as double)")) \
    .withColumn("total_amount", expr("try_cast(total_amount as double)"))
```

```python
from pyspark.sql.functions import col

df_clean = df_typed.filter(
    (col("trip_distance") > 0) &
    (col("fare_amount") > 0) &
    (col("total_amount") > 0)
)

df_clean = df_clean.dropna(subset=[
    "pickup_datetime",
    "hour_of_day",
    "pickup_location_id",
    "dropoff_location_id"
])
```

```python
trips_by_hour = df_clean.groupBy("hour_of_day") \
    .count() \
    .orderBy("hour_of_day")
```

```python
df_clean.groupBy("day_of_week") \    #Peak Days of the Week
    .count() \
    .orderBy("count", ascending=False) \
    .show()
```

```python
df_clean.select(   #pipeline is now correct and stable.
    "pickup_datetime",
    "year",
    "month",
    "day_of_week",
    "hour_of_day"
).show(5)
```

```python
from pyspark.sql.functions import col

df_features = df_clean.withColumn(
    "tip_percentage",
    (col("tip_amount") / col("fare_amount")) * 100
)
```

```python
df_features.printSchema()
```

```python
top_pickups = df_features.groupBy("pickup_location_id") \
    .count() \
    .orderBy("count", ascending=False) \
    .limit(10)

top_pickups.show()
```

```python
top_dropoffs = df_features.groupBy("dropoff_location_id") \
    .count() \
    .orderBy("count", ascending=False) \
    .limit(10)

top_dropoffs.show()
```

```python
revenue_by_payment = df_features.groupBy("payment_type") \   #Revenue by Payment Type
    .agg(avg("total_amount").alias("avg_total_amount"))

revenue_by_payment.show()
```

```python
df_features.groupBy("hour_of_day") \      #Tip Behavior Analysis
    .agg(avg("tip_percentage").alias("avg_tip_percentage")) \
    .orderBy("hour_of_day") \
    .show()
```
